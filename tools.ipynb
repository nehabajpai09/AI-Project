```python
TOOLS = {
    "retrieve_knowledge": {
        "description": "Fetch relevant banking policies or product info",
        "inputs": ["query"]
    },
    "calculate_emi": {
        "description": "Calculate EMI for a loan",
        "inputs": ["principal", "rate", "tenure_months"]
    },
    "escalate_to_human": {
        "description": "Escalate to human support for risky or unclear cases",
        "inputs": ["reason"]
    }
}
```

```python
from retrieval import retrieve
import math

def retrieve_knowledge_tool(query):
    docs = retrieve(query)
    return "\n".join(docs) if docs else "No relevant documents found."


def calculate_emi_tool(principal, rate, tenure_months):
    r = rate / (12 * 100)
    emi = (principal * r * (1 + r)**tenure_months) / ((1 + r)**tenure_months - 1)
    return f"Estimated EMI: ₹{round(emi, 2)}"


def escalate_tool(reason):
    return f"Escalating to human support. Reason: {reason}"
```

```python
def decide_tool(user_input):
    text = user_input.lower()

    if "emi" in text or "loan payment" in text:
        return "calculate_emi"

    if "fraud" in text or "hacked" in text:
        return "escalate_to_human"

    if any(word in text for word in ["charge", "interest", "card", "loan"]):
        return "retrieve_knowledge"

    return None
```

```python
def agent_with_tools(user_input):
    tool_name = decide_tool(user_input)

    # --- No Tool Case ---
    if not tool_name:
        return "I'm not sure. Let me connect you to support."

    # --- Tool Execution ---
    try:
        if tool_name == "retrieve_knowledge":
            result = retrieve_knowledge_tool(user_input)
            return f"Based on bank policy:\n{result}"

        elif tool_name == "calculate_emi":
            # Simple parsing (baseline limitation)
            return calculate_emi_tool(500000, 8.5, 60)

        elif tool_name == "escalate_to_human":
            return escalate_tool("Potential fraud detected")

    except Exception as e:
        return "Tool execution failed. Escalating to support."
```
