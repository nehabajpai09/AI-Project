```python
def plan_steps(user_input):
    text = user_input.lower()

    steps = []

    if "emi" in text:
        steps.append("extract_loan_details")
        steps.append("calculate_emi")

    elif "charged" in text or "fee" in text:
        steps.append("retrieve_knowledge")

    elif "fraud" in text or "hacked" in text:
        steps.append("escalate")

    else:
        steps.append("fallback")

    return steps
```
