```python
def agent_v7(user_input):
    prefs = feedback_store.get_preferences()

    tool = adaptive_tool_decision(user_input, prefs)

    if tool == "retrieve_knowledge":
        response = retrieve_knowledge_tool(user_input)

    elif tool == "calculate_emi":
        response = calculate_emi_tool(500000, 8.5, 60)

    elif tool == "escalate_to_human":
        response = escalate_tool("Risk detected")

    else:
        response = "Let me connect you to support."

    # Adapt style
    response = adapt_response_style(response, prefs)

    return response
```
