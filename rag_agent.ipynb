```python
from openai import OpenAI
from retrieval import retrieve

client = OpenAI(api_key="YOUR_API_KEY")


def generate_rag_response(user_query):
    docs = retrieve(user_query)

    if not docs:
        return "⚠️ I couldn't find relevant banking information. Please contact support."

    context = "\n".join(docs)

    prompt = f"""
You are a safe banking support AI.

Use ONLY the provided context to answer.
If answer is not in context, say you don't know.

Context:
{context}

User Query:
{user_query}
"""

    response = client.chat.completions.create(
        model="gpt-5",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2
    )

    return response.choices[0].message.content


if __name__ == "__main__":
    while True:
        query = input("You: ")

        if query == "exit":
            break

        print(generate_rag_response(query))
```
