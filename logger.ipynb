```python
import logging
import time

logging.basicConfig(
    filename="logs/app.log",
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)

def log_request(user_input):
    logging.info(f"REQUEST: {user_input}")

def log_response(response, latency):
    logging.info(f"RESPONSE: {response} | LATENCY: {latency:.2f}s")

def log_error(error):
    logging.error(f"ERROR: {str(error)}")
```
