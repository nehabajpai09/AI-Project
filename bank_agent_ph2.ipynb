```python
import datetime

# --- Safety Rules ---
PROHIBITED_ACTIONS = ["transfer", "send money", "pay", "approve", "transaction"]
SENSITIVE_TOPICS = ["account balance", "my details", "my account number"]

# --- Simple Knowledge Base (Rule आधारित) ---
RESPONSES = {
    "charges": "Bank charges may include maintenance fees, ATM usage fees, or penalties. Please check your account statement or contact support for exact details.",
    "credit card": "Credit cards offer benefits like rewards, cashback, and travel perks. You should compare based on annual fee and usage.",
    "loan interest": "Loan interest rates vary based on loan type, tenure, and credit profile. Please check the bank's official website for current rates.",
    "card declined": "A card may be declined due to insufficient funds, incorrect PIN, or security restrictions. Please contact support if the issue persists.",
}

# --- Logging Function ---
def log_interaction(user_input, response):
    with open("agent_logs.txt", "a") as f:
        f.write(f"{datetime.datetime.now()} | USER: {user_input} | BOT: {response}\n")


# --- Core Agent Logic ---
def generate_response(user_input):
    user_input_lower = user_input.lower()

    # 1. Safety: Block transactions
    for action in PROHIBITED_ACTIONS:
        if action in user_input_lower:
            return "I cannot assist with transactions or money movement. Please use the official banking app or contact support."

    # 2. Safety: Sensitive data
    for topic in SENSITIVE_TOPICS:
        if topic in user_input_lower:
            return "I cannot access or display personal account information. Please check your secure banking app."

    # 3. Basic FAQ matching
    for keyword, response in RESPONSES.items():
        if keyword in user_input_lower:
            return response

    # 4. Escalation fallback
    return "I'm not fully sure about this. Let me connect you to a human support agent."


# --- CLI Loop ---
def run_agent():
    print("=== AI Banking Support Agent (Baseline) ===")
    print("Type 'exit' to quit.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        response = generate_response(user_input)
        print("Agent:", response)

        log_interaction(user_input, response)


if __name__ == "__main__":
    run_agent()
```
