```python
class Memory:
    def __init__(self):
        self.history = []  # short-term
        self.user_profile = {}  # long-term

    def add_to_history(self, user, bot):
        self.history.append({"user": user, "bot": bot})

        # Keep last 5 turns only
        if len(self.history) > 5:
            self.history.pop(0)

    def get_context(self):
        context = ""
        for h in self.history:
            context += f"User: {h['user']}\nBot: {h['bot']}\n"
        return context

    def update_profile(self, key, value):
        self.user_profile[key] = value

    def get_profile(self):
        return self.user_profile

    def reset(self):
        self.history = []
```
