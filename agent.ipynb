```python
import time
from planner import plan_steps
from tools import calculate_emi_tool, escalate_tool
from retrieval import retrieve
from logger import log_request, log_response, log_error


def run_agent(user_input):
    start_time = time.time()
    log_request(user_input)

    try:
        steps = plan_steps(user_input)
        response_parts = []

        for step in steps:

            if step == "calculate_emi":
                response_parts.append(
                    calculate_emi_tool(500000, 8.5, 60)
                )

            elif step == "retrieve_knowledge":
                docs = retrieve(user_input)
                if docs:
                    response_parts.append("\n".join(docs))
                else:
                    response_parts.append(
                        "No verified info found. Please contact support."
                    )

            elif step == "escalate":
                return escalate_tool("High-risk case detected")

            else:
                response_parts.append("Support will assist you.")

        final_response = "\n".join(response_parts)

    except Exception as e:
        log_error(e)
        final_response = "System is down. Please try later."

    latency = time.time() - start_time
    log_response(final_response, latency)

    return final_response
```
