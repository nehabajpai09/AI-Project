```python
class FeedbackStore:
    def __init__(self):
        self.logs = []
        self.preferences = {
            "prefer_detailed": False,
            "avoid_escalation": False,
            "prefer_tools": True
        }

    def add_feedback(self, user_query, response, feedback):
        self.logs.append({
            "query": user_query,
            "response": response,
            "feedback": feedback
        })

        self.update_preferences(feedback)

    def update_preferences(self, feedback):
        # Simple adaptive rules
        if feedback == "negative":
            self.preferences["prefer_detailed"] = True

        elif feedback == "positive":
            self.preferences["prefer_tools"] = True

    def get_preferences(self):
        return self.preferences
```
